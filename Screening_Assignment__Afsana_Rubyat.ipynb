```python

house_data['date'] = pd.to_datetime(house_data['date']) #convert the date column to a proper datetime type
house_data['date']
```

Creating the derived features: total area, house age, and price per square feet to analysis the dataset.

```python
house_data['total_area'] =house_data['sqft_living'] +house_data['sqft_lot']
```

```python
house_data['house_age'] = 2025 - house_data['yr_built']
house_data['price_per_sqft'] = house_data['price'] / house_data['sqft_living']
```

Identifying  and handling outliers. (Here I have take help from AI but I have implemented it by thorough understanding.)

```python

# 1st made a copy of the original dataset
house_data_cleaned = house_data.copy()

# Then I manually remove logically impossible values
house_data_cleaned = house_data[
    (house_data['bathrooms'] > 0) &
    (house_data['sqft_living'] > 0) &
    (house_data['price'] > 0)
]

# and make only the numeric columns where outliers make sense to remove
numeric_cols = ['price', 'bathrooms', 'sqft_living',
                'sqft_lot', 'floors', 'sqft_above']

z_scores = np.abs(stats.zscore(house_data_cleaned[numeric_cols]))
# I have calculated the Z-scores for every column price,sqft_living,total_are etc)
# Define threshold for Z-score
threshold = 3

#The z-score refers to a score the difference of
#standard deviations of a value from the mean value of that column.
# If a value has a Z-score greater than 3, the value is far away from the mean and could be considered an outlier.

# Keep only rows where all Z-scores are below the threshold
filter_house_data = house_data_cleaned[(z_scores < threshold).all(axis=1)]




```

Task 6. Basic Statistical Analysis:

```python
mean_price = filter_house_data['price'].mean()
median_price = filter_house_data['price'].median()
mode_price = filter_house_data['price'].mode()[0]

mean_area = filter_house_data['total_area'].mean()
median_area = filter_house_data['total_area'].median()
mode_area = filter_house_data['total_area'].mode()[0] #mode_area represents the most common total area value in your dataset â€”
                      #the area size that appears most frequently among all the houses.many houses have 1500 sq ft, then mode_area would be 1500.

print(f"Price - Mean: {mean_price}, Median: {median_price}, Mode: {mode_price}")

print(f"Total Area - Mean: {mean_area}, Median: {median_area}, Mode: {mode_area}")


#Price - Mean: 540088.1417665294, Median: 450000.0, Mode: 350000.0
#Total Area - Mean: 17186.867302086706, Median: 9575.0, Mode: 5940
```

```python
from scipy.stats import ttest_ind

group1 = filter_house_data[filter_house_data['bedrooms'] == 2]['price']
group2 = filter_house_data[filter_house_data['bedrooms'] == 4]['price']

t_stat, p_value = ttest_ind(group1, group2, equal_var=False) #performs a 2 sample t-test, for the means of two independent samples.
print("T-statistic Result:", t_stat, "P-value:", p_value)

  #The t-statistic value of -37.30 means
 #a very large difference between the mean prices of
 #houses with 2 bedrooms and 4 bedrooms. The negative sign
 #refers that the mean price of houses with 2 bedrooms is significantly lower than houses with 4 bedrooms
 #P value indicates very small than 0.05 This indicates that the difference between the two groups is statistically significant.
  #so we can reject the hypothesis that there is no difference in the mean prices of houses with 2 and 4 bedrooms.
```

```python
waterfront_p = filter_house_data[filter_house_data['waterfront'] == 1]['price']
non_waterfront_p = filter_house_data[filter_house_data['waterfront'] == 0]['price']

t_stat, p_value = stats.ttest_ind(waterfront_p, non_waterfront_p)
print(f"T-statistic: {t_stat}, P-value: {p_value}")


#T statistic the large value suggests  the mean price of waterfront houses is much higher
# than that of nonwaterfront houses.As the p-value is much smaller than thresholds (like 0.05),
#wecan reject the null hypothesis that there is no difference in the mean prices between the two groups.
```

Waterfront properties have higher prices. The median price is higher than non-waterfront properties. The distribution is wider, meaning more variability in prices, meaning not all of them are extremely costly.

Non-waterfront properties  have a lower and more concentrated price range.Prices are more tightly clustered below 1. A few properties exceed 1.5, but they are rare.

Some very expensive non-waterfront homes exist. Some cheaper waterfront properties also exist, but they are fewer.
